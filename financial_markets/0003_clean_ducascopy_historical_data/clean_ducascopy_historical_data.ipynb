```python
import pandas as pd
```

```python
df = pd.read_csv('GBPUSD_Candlestick_1_Hour_BID_01.01.2010-11.10.2025.csv')
```

```python
df['datetime'] = df['Local time'].apply(lambda x: x.split(' ')[0] + ' ' + x.split(' ')[1])
```

```python
df['datetime'] = pd.to_datetime(df['datetime'])
```

```python
import datetime as dt

df['timestamp'] = df['datetime'].apply(lambda x: x.timestamp())
```

```python
df['timestamp'] = df['timestamp'].astype(int)
```

```python
df = df[df['Volume'] > 0]
```

```python
df = df.rename(
    columns={
        'Open' : 'open',
        'High' : 'high',
        'Low' : 'low',
        'Close' : 'close',
        'Volume' : 'volume',
    }
)
```

```python
df[['timestamp', 'open', 'high', 'low', 'close', 'volume']].to_excel('gbp_usd_1h_98413.xlsx')
```
